 Practice 2 - Chapter 2: Build a model which predicts sales based on the money spent on different platforms for marketing

```
import pandas as pd
import numpy as np

theta_0 = 0 #define the bias of Hypothesis function

# Read CSV file
file_path = "/content/Practice2_Chapter2.csv"
df = pd.read_csv(file_path)

# export value of data to x and y
x = df[['TV', 'Radio', 'Newspaper']].values
y = df['Sales'].values
print("Giá trị của data:", x)
print("Giá trị của output:\n", y)
```

```
# Feature scaling
# This function is using to determine the scale of each column (for each feature of data)
def feature_scaling(x):
    mean_val = np.mean(x, axis=0) # axis = 0 means evaluate value of each column
    max_val = np.max(x, axis=0)
    x_scaled = np.abs((x - mean_val) / max_val)
    return x_scaled, mean_val, max_val

x_scaled, mean_val, max_val = feature_scaling(x)
print(f'giá trị x scaled {x_scaled}')
print(f'giá trị mean {mean_val}')


# Initialize parameters
theta = np.zeros(x_scaled.shape[1])  # [theta_0, theta_1, ..., theta_n]
learning_rate = 0.01
n = 5000 # n: number of interactions

def hypothesis(x, theta):
    return np.dot(x, theta) + theta_0

def cost_function(x, y, theta):
    m = len(y)
    predictions = hypothesis(x, theta)
    return (1 / (2 * m)) * np.sum((predictions - y) ** 2)

def gradient_descent(x, y, theta, learning_rate, n):
    m = len(y)
    history = {"theta": [], "cost": []}

    for i in range(n):
        # Compute gradients
        predictions = hypothesis(x, theta)
        gradients = (1 / m) * np.dot(x.T, (predictions - y))

        # Update theta
        theta = theta - learning_rate * gradients

        # Save cost and theta for tracking
        cost = cost_function(x, y, theta)
        history["theta"].append(theta)
        history["cost"].append(cost)


    return theta, history

# Train the model using gradient descent
theta, history = gradient_descent(x_scaled, y, theta, learning_rate, n)

# Make predictions
predicted_sales = hypothesis(x_scaled, theta)

# Inverse scaling
def inverse_scaling(theta, mean_val, max_val):
    # Ensure mean_val and max_val are sliced to match features
    mean_val = mean_val[:len(theta)-1]
    max_val = max_val[:len(theta)-1]

    # Compute theta_original
    theta_original = np.zeros_like(theta)
    theta_original[0] = theta[0] - np.sum((theta[1:] * mean_val) / max_val)
    theta_original[1:] = theta[1:] / max_val

    return theta_original

theta_original = inverse_scaling(theta, mean_val, max_val)

# Print results
print(f"Final scaled theta: {theta}")
print(f"Final original theta: {theta_original}")
print(f"Predicted sales: {predicted_sales}")

```
